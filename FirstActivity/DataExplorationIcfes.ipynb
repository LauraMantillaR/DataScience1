Which do you think are the most important columns? If we are looking at the overall score.

```python
correlation_matrix = df.corr()
print(correlation_matrix) #Square matrix showing the correlation between all columns.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Create heat map
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
#Show the graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
#Select non-numeric columns
no_numeric_columns = df.select_dtypes(exclude=['int', 'float']).columns
#Encode non-numeric columns with LabelEncoder
le = LabelEncoder()
for column in no_numeric_columns:
    df[column] = df[column].astype(str) #Change the values in the current column to strings using the astype() method. This is necessary because LabelEncoder() requires inputs to be strings or numbers.
    df[column] = le.fit_transform(df[column]) #Applies the label encoding transformation to the column using the fit_transform() method.  The transformed values are then stored back in the same column of the DataFrame. 
#Correlation matrix 2
correlation_matrix2 = df.corr()
print(correlation_matrix2)
#Create heat map
sns.heatmap(correlation_matrix2, annot=True, cmap='coolwarm')
#Show the graph
plt.show()
```
