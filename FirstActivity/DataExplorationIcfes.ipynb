Which do you think are the most important columns? If we are looking at the overall score.

```python
correlation_matrix = df.corr()
print(correlation_matrix) #Square matrix showing the correlation between all columns.
```

Correlations:

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Create heat map
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
#Show the graph
plt.show()
```

```python
#Select columns of interest
df_subset = df[["PUNT_GLOBAL", "FAMI_ESTRATOVIVIENDA"]]

# Agrupar por departamento y calcular la media de los puntajes globales
df_grouped = df_subset.groupby(["FAMI_ESTRATOVIVIENDA"]).mean().reset_index()

# Crear diagrama de caja y bigotes
plt.figure(figsize=(12,6))
sns.boxplot(x="FAMI_ESTRATOVIVIENDA", y="PUNT_GLOBAL", data=df_subset, width=0.5)
plt.title("Puntajes globales según estrato")
plt.xlabel("Estrato Vvienda")
plt.ylabel("Puntaje global")
plt.show()
```

```python
#Select columns of interest
df_subset = df[["PUNT_GLOBAL", "COLE_AREA_UBICACION"]]

#Group by the area where the school is located and calculate the mean of the overall scores
df_grouped = df_subset.groupby(["COLE_AREA_UBICACION"]).mean().reset_index()

#Create box-and-whisker plot
plt.figure(figsize=(12,6))
sns.boxplot(x="COLE_AREA_UBICACION", y="PUNT_GLOBAL", data=df_subset, width=0.5)
plt.title("Global scores by area")
plt.xlabel("Area")
plt.ylabel("Global score")
plt.show()
```

```python
#Select columns of interest
df_subset = df[["PUNT_GLOBAL", "COLE_JORNADA"]]

#Group by school day and calculate the mean of the overall scores
df_grouped = df_subset.groupby(["COLE_JORNADA"]).mean().reset_index()

#Create box-and-whisker plot
plt.figure(figsize=(12,6))
sns.boxplot(x="COLE_JORNADA", y="PUNT_GLOBAL", data=df_subset, width=0.5)
plt.title("Global scores according to school day")
plt.xlabel("Schoold day")
plt.ylabel("Global score")
plt.show()
```

```python
#Select columns of interest
df_subset = df[["PUNT_GLOBAL", "ESTU_GENERO"]]

#Group by gender and calculate the mean of the overall scores
df_grouped = df_subset.groupby(["ESTU_GENERO"]).mean().reset_index()

#Create box-and-whisker plot
plt.figure(figsize=(12,6))
sns.boxplot(x="ESTU_GENERO", y="PUNT_GLOBAL", data=df_subset, width=0.5)
plt.title("Global scores by gender")
plt.xlabel("Gender")
plt.ylabel("Global score")
plt.show()
```
